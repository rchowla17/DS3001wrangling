```python
#Cleaning type variable from sharks.csv

df = pd.read_csv('./data/sharks.csv')
df.head()
```

```python
import numpy as np


types = df['Type']

types = types.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft') #all water related
types.value_counts()

types = types.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # na
types.value_counts()

df['Type'] = types # Replace the 'Type' variable with the cleaned version
del types # Destroy the temporary vector

df['Type'].value_counts()
```

## Part 4: Missing values, not at random:

For the pretrial data covered in the lecture, clean the ImposedSentenceAllChargeInContactEvent variable as well as you can, and explain the choices you make. (Hint: Look at the SentenceTypeAllChargesAtConvictionInContactEvent variable.)

```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

length = pd.to_numeric(length,errors='coerce') # Coerce to numeric
length_NA = length.isnull()
print( "Missing: ", np.sum(length_NA),'\n') #too many missing

print( pd.crosstab(length_NA, type), '\n')

length = length.mask( type == 4, 0)  #category 4 is released (sentence length 0)
length = length.mask( type == 9, np.nan)

length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( "New missing: ", np.sum(length_NA),'\n')

df['ImposedSentenceAllChargeInContactEvent'] = length # Replace data with cleaned version
del length, type # Delete temporary length/type variables
```
